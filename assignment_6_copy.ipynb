### 1.1a. $ (A \backslash B) \cup (A \cap B) = A $

This says that the union (denoted $\cup$) of the relative complement of sets A and B (denoted $(A \backslash B)$) and the intersection of sets A and B (denoted $(A \cap B)$), is equal to A.

And we know that:
- The union of two sets is the set of the elements that are elements of at least one of the sets.<br>So, $A \cup B = \{x: x \in A \text{ or } x \in B \}$.

- The relative complement (sometimes called the *difference*) of two sets $A$ and $B$ is the set of elements which are elements that belong to $A$ but not to $B$.<br>So, $A \backslash B = \{x: x \in A \text{ and } x \notin B \}$.

- The intersection of two sets is the set made up of the elements which are elements of both sets.<br>So, $A \cap B = \{x: x \in A \text{ and } x \in B \}$.


Before our proof, we can understand this intuitively: We can say that it makes sense that when you combine the part of A that isn't in B (the relative complement) and the part of A that is in B (the intersection), you get all of A.

Now we will prove this element wise by using double inclusion to show two sets equal.

To prove two sets $X$ and $Y$ are equal, we have to show that:
- $X \subseteq Y$ (every element in $X$ is also in $Y$), and
- $Y \subseteq X$ (every element in $Y$ is also in $X$)
#### **Theorem**: For any two sets $A$ and $B$, $(A \backslash B) \cup (A \cap B) = A$.
#### **Proof**:

##### **Part I**: Prove $(A \backslash B) \cup (A \cap B) \subseteq A$
**Goal**: We must show that any element in $(A \backslash B) \cup (A \cap B)$ is also in $A$.

Let $x$ be an arbitrary element such that $x \in (A \backslash B) \cup (A \cap B)$.

By the definition of a union, we know that $x \in (A \backslash B) \cup (A \cap B)$ means that $x \in (A \backslash B)$ **OR** $x \in (A \cap B)$.

- Case 1: Assume $x \in (A \backslash B)$.
    - By definition of the relative complement, $x \in A$ and $x \notin B$.

- Case 2: Assume $x \in (A \cap B)$.
    - By definition of the intersection, $x \in A$ and $x \in B$.

In both Case 1 and Case 2, $x \in A$. Since $x$ is an arbitrary element, we've shown that any element of $(A \backslash B) \cup (A \cap B)$ is also an element of $A$.

Therefore, by the definition of a subset ($X$ is a subset of $Y$ if every element of $X$ is also an element of $Y$), we've proven that $(A \backslash B) \cup (A \cap B) \subseteq A$.

##### **Part II**: Prove $A \subseteq (A \backslash B) \cup (A \cap B)$
**Goal**: Now we must show that any element in $A$ is also in $(A \backslash B) \cup (A \cap B)$.

Let $x$ be an arbitrary element such that $x \in A$.

Consider any element $x$ and any set $B$. By the Law of the Excluded Middle (who knew that had a formal name? not me until today), there are two mutually exclusive possibilities: either $x \in B$ is true or $x \notin B$ is true. [I was going to do a proof by contradiction for this proof section, but did some research and realized this was a more direct method].

- Case 1: Assume $x \in B$.
    - $x \in A$ and $x \in B$. By definition of the intersection, $x \in (A \cap B)$.

- Case 2: Assume $x \notin B$.
    - $x \in A$ and $x \notin B$. By definition of the relative complement, $x \in (A \backslash B)$.

So, from our two cases we have shown that, since $x \in A$, $x \in (A \cap B)$ **or** $x \in (A \backslash B)$. By the definition of a union, this means $x \in (A \backslash B) \cup (A \cap B)$.

Since $x$ is an arbitrary element, we've shown that any element of $A$ is also an element of $(A \backslash B) \cup (A \cap B)$.

Therefore, by the definition of a subset, we've proven that $A \subseteq (A \backslash B) \cup (A \cap B)$.

We've now proven that $(A \backslash B) \cup (A \cap B) \subseteq A$ and that $A \subseteq (A \backslash B) \cup (A \cap B)$. 

By the definition of set equality, $(A \backslash B) \cup (A \cap B) = A$. $\square$

### 1.1b. $ A \cup ( B \cap C) = (A \cup B) \cap (A \cup C)$

This is the distributive law of union over intersection.

#### **Theorem**: For any three sets $A$, $B$, and $C$, $A \cup ( B \cap C) = (A \cup B) \cap (A \cup C)$.
#### **Proof**:

##### **Part I**: Prove $A \cup ( B \cap C) \subseteq (A \cup B) \cap (A \cup C)$
**Goal**: We must show that any element in $A \cup ( B \cap C)$ is also in $ (A \cup B) \cap (A \cup C)$.

Let $x$ be an arbitrary element such that $x \in A \cup ( B \cap C)$.

By the definition of a union, we know that $x \in A \cup (B \cap C)$ means that $x \in A$ **OR** $x \in (B \cap C)$.

- Case 1: Assume $x \in A$.
    - By definition of the union, if $x \in A$, then $x \in A \cup K$ where $K$ is any arbitrary set.

    - So, $x \in (A \cup B)$ **AND** $x \in (A \cup C)$.
    - By definition of the intersection, as **x** is in both sets, it is in their intersection.
        - So $x \in (A \cup B) \cap (A \cup C)$.

- Case 2: Assume $x \in (B \cap C)$.
    - By definition of the intersection, if $x \in (B \cap C)$, then $x \in B$ **AND** $x \in C$.

    - By definition of the union, if $x \in B$, then $x \in (A \cup B)$.
    - Similarly, if $x \in C$, then $x \in (A \cup C)$.
    - And by definition of the intersection, as **x** is in both sets, it is in their intersection.
        - So $x \in (A \cup B) \cap (A \cup C)$.


In both Case 1 and Case 2, $x \in (A \cup B) \cap (A \cup C)$. Since $x$ is an arbitrary element, we've shown that any element of $A \cup ( B \cap C)$ is also in $ (A \cup B) \cap (A \cup C)$.

Therefore, by the definition of a subset ($X$ is a subset of $Y$ if every element of $X$ is also an element of $Y$), we've proven that $A \cup ( B \cap C) \subseteq (A \cup B) \cap (A \cup C)$.

##### **Part II**: Prove $(A \cup B) \cap (A \cup C) \subseteq A \cup ( B \cap C) $
**Goal**: We must show that any element in $ (A \cup B) \cap (A \cup C) $ is also in $ A \cup ( B \cap C)$.

Let $x$ be an arbitrary element such that $x \in (A \cup B) \cap (A \cup C) $.

By the definition of an intersection, we know that $x \in (A \cup B) \cap (A \cup C) $ means that $x \in (A \cup B)$ **AND** $x \in (A \cup C)$.

Consider $x$ and set $A$.

- Case 1: Assume $x \in A$.
    - By definition of the union, if $x \in A$, then $x \in A \cup K$ where $K$ is any arbitrary set.

    - So, $x \in A \cup (B \cap C)$.

- Case 2: Assume $x \notin A$.
    - We know (see our above statement about the definition of an intersection), that $x \in (A \cup B)$ **AND** $x \in (A \cup C)$.

    - Regarding $(A \cup B)$:
        - We know from the definition of the union that if $ x \notin A$, it must be so that $x \in B$.
    - Regarding $(A \cup C)$:
        - We know from the definition of the union that if $ x \notin A$, it must be so that $x \in C$.
    - Therefore, $x$ must be in $B$ **AND** $C$.
    - So, by definition of the intersection, $x \in (B \cap C)$.
    - And, consequently by definition of the union, it is such that $x \in A \cup (B \cap C)$.

In both Case 1 and Case 2, $x \in A \cup (B \cap C)$. Since $x$ is an arbitrary element, we've shown that any element of $ (A \cup B) \cap (A \cup C) $ is also in $ A \cup ( B \cap C)$.

Therefore, by the definition of a subset ($X$ is a subset of $Y$ if every element of $X$ is also an element of $Y$), we've proven that $(A \cup B) \cap (A \cup C) \subseteq A \cup ( B \cap C) $.

We've now proven that $A \cup ( B \cap C) \subseteq (A \cup B) \cap (A \cup C)$ and that $(A \cup B) \cap (A \cup C) \subseteq A \cup ( B \cap C) $. 

By the definition of set equality, $A \cup ( B \cap C) = (A \cup B) \cap (A \cup C)$. $\square$

### 1.1c. $ A \backslash (B \cup C) = A\backslash B \cap A \backslash C $ and $A \backslash (B \cap C) = A \backslash B \cup A \backslash C$

These are actually versions of De Morgan's Laws.

This is really two proofs.
- Theorem 1: $A \backslash (B \cup C) = A\backslash B \cap A \backslash C$
- Theorem 2: $A \backslash (B \cap C) = A \backslash B \cup A \backslash C$

#### **Theorem 1**: For any three sets $A$, $B$, and $C$, $A \backslash (B \cup C) = A\backslash B \cap A \backslash C$.

#### **Proof**: 

##### **Part I**: Prove $A \backslash (B \cup C) \subseteq A\backslash B \cap A \backslash C$
**Goal**: We must show that any element in $A \backslash ( B \cup C)$ is also in $A \backslash B \cap A \backslash C$.

Let $x$ be an arbitrary element such that $x \in A \backslash ( B \cup C)$.

By the definition of the relative complement, we know that $x \in A \backslash ( B \cup C)$ means that $x \in A$ **AND** $x \notin (B \cup C)$.

By definition of the union, since $x \notin (B \cup C)$, we know that $x \notin B$ **AND** $x \notin C$. 

So, $x \in (A \backslash B)$ **AND** $x \in (A \backslash C)$.

By the definition of the intersection, as $x$ is in both relative complements, it is the case that $x \in A \backslash B \cap A \backslash C$.

Since $x$ is an arbitrary element, we've shown that any element of $A \backslash ( B \cup C)$ is also in $A \backslash B \cap A \backslash C$.

Therefore, by the definition of a subset ($X$ is a subset of $Y$ if every element of $X$ is also an element of $Y$), we've proven that $A \backslash (B \cup C) \subseteq A\backslash B \cap A \backslash C$.

##### **Part II**: Prove $A \backslash B \cap A \backslash C \subseteq A \backslash (B \cup C)$
**Goal**: We must show that any element in $A \backslash B \cap A \backslash C$ is also in $A \backslash (B \cup C)$.

Let $x$ be an arbitrary element such that $x \in A \backslash B \cap A \backslash C$.

By the definition of an intersection, we know that $x \in A \backslash B \cap A \backslash C$ means that $x \in A \backslash B$ **AND** $x \in A \backslash C$.

By definition of the relative complement, since $x \in A \backslash B$, we know $x \in A$ and $x \notin B$.

Similarly, since $x \in A \backslash C$, we know $x \in A$ and $x \notin C$.

Assume, for the sake of contradiction, that $x \in (B \cup C)$.
- By definition of a union, this means $x \in B$ **OR** $x \in C$.
- But we know that $x \notin B$ and $x \notin C$. So we have a contradiction.
    - We thus can conclude that $x \notin (B \cup C)$.

As we know $x \in A$ and we now know $x \notin (B \cup C)$, by the definition of the relative complement we can say that $x \in A \backslash (B \cup C)$.

Since $x$ is an arbitrary element, we've shown that any element of $A \backslash B \cap A \backslash C$ is also in $A \backslash (B \cup C)$.

Therefore, by the definition of a subset ($X$ is a subset of $Y$ if every element of $X$ is also an element of $Y$), we've proven that $A \backslash B \cap A \backslash C \subseteq A \backslash (B \cup C)$.

We've now proven that $A \backslash (B \cup C) \subseteq A\backslash B \cap A \backslash C$ and that $A \backslash B \cap A \backslash C \subseteq A \backslash (B \cup C)$. 

By the definition of set equality, $A \backslash (B \cup C) = A\backslash B \cap A \backslash C$. $\square$

#### **Theorem 2**: For any three sets $A$, $B$, and $C$, $A \backslash (B \cap C) = A \backslash B \cup A \backslash C$.

#### **Proof**: 

##### **Part I**: Prove $A \backslash (B \cap C) \subseteq A\backslash B \cup A \backslash C$
**Goal**: We must show that any element in $A \backslash ( B \cap C)$ is also in $A \backslash B \cup A \backslash C$.

Let $x$ be an arbitrary element such that $x \in A \backslash ( B \cap C)$.

By the definition of the relative complement, we know that $x \in A \backslash ( B \cap C)$ means that $x \in A$ **AND** $x \notin (B \cap C)$.

- Case 1: Assume $x \notin B$.

    - By definition of the union, since $x \in A \backslash B$, it is such that $x \in A \backslash B \cup A \backslash C$.

- Case 2: Assume $x \notin C$.
    - Since $x \in A$ and $x \notin C$, by definition of the relative complement, $x \in A \backslash C$.
    - By definition of the union, since $x \in A \backslash C$, it is such that $x \in A \backslash B \cup A \backslash C$.


In both Case 1 and Case 2, $x \in A \backslash B \cup A \backslash C$. Since $x$ is an arbitrary element, we've shown that any element of $A \backslash ( B \cap C)$ is also in $A \backslash B \cup A \backslash C$.

Therefore, by the definition of a subset ($X$ is a subset of $Y$ if every element of $X$ is also an element of $Y$), we've proven that $A \backslash (B \cap C) \subseteq A\backslash B \cup A \backslash C$.


##### **Part II**: Prove $A\backslash B \cup A \backslash C \subseteq A \backslash (B \cap C)$

By the definition of the union, we know that $x \in A\backslash B \cup A \backslash C$ means that $x \in A\backslash B$ **OR** $x \in A \backslash C$.

- Case 1: Assume $x \in A \backslash B$.
    - By definition of the relative complement, since $x \in A \backslash B$, we know $x \in A$ and $x \notin B$.

- Case 2: Assume $x \in A \backslash C$.

Since $x$ is an arbitrary element, we've shown that any element of $A\backslash B \cup A \backslash C$ is also in $A \backslash (B \cap C)$.

We've now proven that $A \backslash (B \cap C) \subseteq A\backslash B \cup A \backslash C$ and that $A\backslash B \cup A \backslash C \subseteq A \backslash (B \cap C)$.

```python
import numpy as np
import matplotlib.pyplot as plt

# Set up the figure with 4 subplots
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
plt.subplots_adjust(hspace=0.3, wspace=0.3) # this is the spacing between subplots

# 1.2a: Set A (1D Number Line)
# Rule: x^2 - 1 >= 0
ax = axes[0, 0] # this selects the top left subplot to work in
x = np.linspace(-3, 3, 400)
y_parabola = x**2 - 1

# Plot the parabola for context (dashed line)
ax.plot(x, y_parabola, 'k--', alpha=0.3, label='$y = x^2 - 1$')
ax.axhline(0, color='black', linewidth=1) # The number line

# Highlight the actual set (Where the parabola is above the line)
# Logic: x <= -1 OR x >= 1
ax.plot([-3, -1], [0, 0], color='blue', linewidth=4, label='Set A')
ax.plot([1, 3], [0, 0], color='blue', linewidth=4)
ax.scatter([-1, 1], [0, 0], color='blue', s=100, zorder=5) # Closed dots for ">="

ax.set_title(r'1.2a: $A = \{x \in \mathbb{R}: x^2 - 1 \geq 0\}$')
ax.set_xlabel('x')
ax.legend()
ax.grid(True, alpha=0.3)


# Setup for 2D Plots (B, C, D)
# Create a grid of points
x_vals = np.linspace(-5, 5, 400)
y_vals = np.linspace(-5, 5, 400)
X, Y = np.meshgrid(x_vals, y_vals)

# 1.2b: Set B (The Half-Plane)
# Rule: 3x - 2y >= 0  (or y <= 1.5x)
ax = axes[0, 1]
mask_b = (3*X - 2*Y >= 0)

# contourf fills the region where the mask is True (1)
ax.contourf(X, Y, mask_b, levels=[0.5, 1.5], colors=['darkblue'], alpha=0.6)
ax.plot(x_vals, 1.5*x_vals, 'b--', linewidth=1) # Boundary line

ax.set_title(r'1.2b: $B = \{3x - 2y \geq 0\}$')
ax.set_xlabel('x'); ax.set_ylabel('y')
ax.set_xlim(-5, 5); ax.set_ylim(-5, 5)
ax.grid(True, alpha=0.3)
ax.axhline(0, color='k', linewidth=0.5); ax.axvline(0, color='k', linewidth=0.5)


# 1.2c: Set C (The Hyperbolas)
# Rule: xy >= 3
ax = axes[1, 0]
mask_c = (X * Y >= 3)

ax.contourf(X, Y, mask_c, levels=[0.5, 1.5], colors=['purple'], alpha=0.6)
# Plot standard hyperbola lines for clarity
ax.plot(x_vals, 3/x_vals, 'purple', linestyle='--', alpha=0.5)

ax.set_title(r'1.2c: $C = \{xy \geq 3\}$')
ax.set_xlabel('x'); ax.set_ylabel('y')
ax.set_xlim(-5, 5); ax.set_ylim(-5, 5)
ax.grid(True, alpha=0.3)
ax.axhline(0, color='k', linewidth=0.5); ax.axvline(0, color='k', linewidth=0.5)


# 1.2d: Set D (this is the Simplex)
# Rule: x1 >= 0, x2 >= 0, x1 + x2 <= 1
# This shape is called a "Simplex" - (crucial in optimization)
ax = axes[1, 1]
mask_d = (X >= 0) & (Y >= 0) & (X + Y <= 1)

ax.contourf(X, Y, mask_d, levels=[0.5, 1.5], colors=['lightblue'], alpha=0.6)

ax.set_title(r'1.2d: $D = \{x_1, x_2 \geq 0, x_1 + x_2 \leq 1\}$')
ax.set_xlabel('$x_1$'); ax.set_ylabel('$x_2$')
ax.set_xlim(-1, 2); ax.set_ylim(-1, 2)
ax.grid(True, alpha=0.3)
ax.axhline(0, color='k', linewidth=0.5); ax.axvline(0, color='k', linewidth=0.5)

plt.tight_layout()
plt.show()
```

### Imagine rolling a fair single six-sided die. There are 6 outcomes, all equally likely. Derive the sample space and the space of events. What are the probabilities of the outcomes and events? 
The sample space (the set of all possible outcomes) is $\set{1,2,3,4,5,6}$.

The space of events (the set of all possible events), is the power set of the sample space. This is the set of all subsets of the sample space (including the empty set and the entire sample space itself). There are $2^6=64$ events in the space of events.
- $\set{}$, probability is zero
- $\set{ 1, 2, 3, 4, 5,6 }$, probability is 1/6
- $\set{ 12, 13, 14, 15, 16, 23, 24, 25, 26, 34, 35, 36, 45, 46, 56 }$, probability is 2/6
- $\set{ 123, 124, 125, 126, 134, 135, 136, 145, 146, 156, 234, 235, 236, 245, 246, 256, 345, 346, 356, 456 }$, probability is 3/6
- $\set{1234, 1235, 1246, 1236, 1245, 1246, 1256, 1345, 1346, 1356, 1456, 2345, 2346, 2356, 2456, 3456}$, probability is 4/6
- $\set{12345, 12346, 12356, 12456, 13456, 23456 }$, probability is 5/6
- $\set{123456}$, probability is 1

Since the die is fair, all 6 outcomes are equally likely. As all probabilities must sum to 1, the probability of one single outcome is $1/6$. So, if we call that outcome $z$, $$P(z)=P(1)=P(2)= \dots = P(6)=1/6.$$

The probability of an event is the sum of the probabilities of all the outcomes in that event.
- $P(\text{rolling a 5})=P(\set{5})=1/6$

- $P(\text{rolling an odd number})=P(\set{1,3,5})=P(1)+P(3)+P(5)=1/6 + 1/6 + 1/6 = 3/6 = 1/2$

### 4.1 Write code to model spinning the roulette wheel, including the colors and numbers (you could make two lists of number and color and draw a random number between 0 and 37... or use a dataframe with color and number variables and sample it... or use a dict with key to number/color pairs...)

```python
import numpy as np
import pandas as pd

# define the 18 red numbers
red_numbers = {1, 3, 5, 7, 9, 12, 14, 16, 18, 19, 21, 23, 25, 27, 30, 32, 34, 36}
numbers = list(range(37)) # all numbers from 0 to 36
colors = []
is_odd = []

# assign a color to each number
for num in numbers:
    if num == 0:
        colors.append('Green')
        is_odd.append(False) # in roulette, 0 is neither odd nor even for betting
    elif num in red_numbers:
        colors.append('Red')
        is_odd.append(num % 2 != 0)
    else:
        # if its 1-36 and not red, its black
        colors.append('Black')
        is_odd.append(num % 2 != 0)

# create wheel data frame
roulette_wheel = pd.DataFrame({
    'Number': numbers,
    'Color': colors,
    'IsOdd': is_odd
})

roulette_wheel.head()
```

```python
# lets spin the wheel by randomly sampling a row of the data frame
def spin_wheel(wheel_df):
    """
    Simulates a spin of the 37 slot European roulette wheel by randomly sampling a row of the data frame.
    Returns a row of the data frame representing the outcome of the spin.
    """
    n_spins = 1
    outcome = wheel_df.sample(n_spins, replace=True) # replace true is critical
    return outcome

# test it out
print(spin_wheel(roulette_wheel))
```

### 4.2 Describe the probability space associated with the roulette wheel: Outcomes, events, probabilities (If there are 37 outcomes, there are $2^{37}= 137,438,953,472$ events, by the way)

Since the wheel is fair, every single slot bet is equally likely. So $P(z)=\frac{1}{37}$ for any single number.

For bets (events) with more than one element, we count up the number of elements and divide by 37 to get the probability.

### 4.3 You wrote code to generate a spin of the roulette wheel. Now write a function that takes a basic or complex bet as an argument, and returns the result for the player (win or lose, and the payout 36/K-1 or -1)

```python
def play_roulette(bet_choice, n_spins=1):
    '''Takes a bet choice which can be:
    A string: 'Black', 'Red', 'Odd', or 'Even'
    A list of numbers for complex bets: [1,2], [1,2,3], etc.
    A single number in a list for basic bets: [1], [2], etc.
    '''

    # 1st step: Determine winning numbers
    if isinstance(bet_choice, list): # if bet_choice is a list
        winning_numbers = bet_choice
    
    elif bet_choice == 'Black':
        # get all black numbers from data frame and put into a list
        winning_numbers = roulette_wheel.loc[roulette_wheel['Color'] == 'Black','Number'].tolist()
    elif bet_choice == 'Red':
        # get all red numbers from data frame and put into a list
        winning_numbers = roulette_wheel.loc[roulette_wheel['Color'] == 'Red','Number'].tolist()
    elif bet_choice == 'Odd':
        # get all odd numbers from data frame and put into a list
        winning_numbers = roulette_wheel.loc[roulette_wheel['IsOdd'] == True,'Number'].tolist()
    elif bet_choice == 'Even':
        # get all even numbers excluding 0 from data frame and put into a list
        winning_numbers = roulette_wheel.loc[
            (roulette_wheel['IsOdd'] == False) & (roulette_wheel['Number'] != 0),
            'Number'].tolist()
    else:
        # if bet_choice is not a list or one of the strings above, raise an error
        raise ValueError(f"Invalid bet: {bet_choice}")
    
    # 2nd step: Calculate payout odds

    K = len(winning_numbers) # how many numbers allow for a win
    if K == 0:
        return "Error: No winning numbers"
    
    payout_if_win = (36/K)-1

    # Step 3: Spin the wheel
    spin_results = roulette_wheel['Number'].sample(n_spins, replace=True).values
    # this is an array of length n_spins

    # check if spin_result is in the list of winning numbers
    did_win = np.isin(spin_results, winning_numbers)

    # create an array of profits, payout if win else -1
    profits = np.where(did_win, payout_if_win, -1)

    # need to return if win or lose and profit as a tuple
    outcomes = np.where(did_win, 'Win', 'Lose')

    return outcomes, profits
```

### Exponential distribution (similar to log-normal):
$$
F_X(x) = \begin{cases}
0, & x<0 \\
1 - e^{-\lambda x}, & x \ge 0,
\end{cases}
$$
with $\lambda > 0$.

#### 6.2a Compute the density from the distribution for the exponential distribution (take a derivative)

To get the PDF, $f(x)$, we take the derivative of the CDF, $F(x)$.

Where $x<0$, $F'(x) = f(x) = 0$ because the derivative of a constant is $0$.

Where $x \geq 0$, $F'(x) = f(x) = \frac{d}{dx}(1-e^{-\lambda x})$.

With the sum/difference rule, we have
$$f(x)=\frac{d}{dx}(1)-\frac{d}{dx}(e^{-\lambda x})$$

Derivative of a constant is 0, so we have
$$f(x)=0-\frac{d}{dx}(e^{-\lambda x})$$

Then we apply chain rule to $e^{-\lambda x}$
- $\frac{d}{du}e^u = \frac{d}{du}e^u * \frac{du}{dx}u$

- Let $u = -\lambda x$, so $\frac{du}{dx} = -\lambda$.

So, we can write
$$f(x)=-(e^{-\lambda x})*(-\lambda)$$

Then, we find
$$f(x)=\lambda e^{-\lambda x}$$

Finally, we can state
$$f(x)= \begin{cases}0, \qquad \quad x<0 \\\lambda e^{-\lambda x}, \quad x \geq 0 \end{cases}$$

#### 6.2b Determine the support

When $x<0$, $f(x) = 0$, and thus such values of $x$ can't be included in the support.

The exponential function $e^x$ can take any real number and will return a nonzero, positive result. Since $\lambda$ is a constant and $e^x$ works for any real input, there's no non-negative value of $x$ that will make $f(x) \leq 0$.

So the support is all non-negative real numbers: $\mathbb{R}_{\geq 0} = [0, +\infty)$

#### 6.3a Determine the support

If we look at $k$ as the number of failures that you have before you reach $r$ successes, we note a few things:
- It is possible to have $0$ failures. (Immediately get $r$ heads flips)
- We are dealing with integers. (Can't have half a failure)
- Can have infinite failures before reaching $r$ successes.

So, our support is $k \in \set{0,1,2,\dots}$ or $\mathbb{Z}_{\geq 0}$, or $\mathbb{N}_0$. (The set of natural numbers / positive integers)

#### 6.3b Plot the density and distribution for a variety of parameter values

The probability mass function gives the probability at a point. The CDF gives the total probability up to a point.

Note: Scipy nbinom defines $p$ as the probability of success but it counts the number of failures. It uses $n$ to count the number of successes.

```python
# step 1: define range of k failures to look at
k_vals = np.arange(0,30)

# step 2: define some (r,p) combinations to compare
# r = number of successes (n in scipy), p = probability of success
params = [(5, 0.5), (10, 0.5), (5, 0.25)]

# step 3: plot the theoretical pdf and cdf
fig, ax = plt.subplots(1,2,figsize=(14,5))

for r,p in params:
    # calculate the pmf and cdf
    pmf = nbinom.pmf(k=k_vals, n=r, p=p)
    cdf = nbinom.cdf(k=k_vals, n=r, p=p)
    
    # plot the pmf (i'll use a line style here for clarity with multiple lines,
    # but technically it's discrete points)
    # ax[0].plot(k_vals, pmf, 'o--', label=f'r={r}, p={p}', alpha=0.6)
    # or as points
    ax[0].scatter(k_vals, pmf, label=f'r={r}, p={p}')
    
    # plot the cdf (using step to show discreteness)
    ax[1].step(k_vals, cdf, where='post', label=f'r={r}, p={p}')

ax[0].set_title('Density')
ax[0].legend()
ax[0].grid(True, alpha=0.3) # add a grid for clarity


ax[1].set_title('Distribution')
ax[1].legend()
ax[1].grid(True, alpha=0.3)

plt.show()
```

#### 6.3c Take a sample of 1000 draws $(x_1, x_2, ..., x_{1000})$ from the distribution, plot the relative frequency and ECDF, visually compare with the theoretical pmf/cdf

```python
# step 1: the simulation (sampling)

# get 1000 samples from the negative binomial distribution
# pick one set of parameters to keep things simple
sim_r = 5       # n in scipy
sim_p = 0.5
n_samples = 1000

# step 2: generate the sample
sample = nbinom.rvs(n=sim_r, p=sim_p, size=n_samples)

# step 3: plot sample relative frequency as histogram vs theoretical PMF as points
fig, ax = plt.subplots(1,2, figsize=(14,5))

# plot the sample relative frequency
# use discrete=True to make seaborn treat the sample as discrete (respect integer bins)
sns.histplot(sample, stat='probability', discrete=True, ax=ax[0], label='Sample Relative Frequency', color='blue', alpha=0.3)

# overlay the theoretical probabilities (as points)
theoretical_pmf = nbinom.pmf(k=k_vals, n=sim_r, p=sim_p)
ax[0].plot(k_vals, theoretical_pmf, 'ro', # marker='o', color='red',
            label = 'Theoretical PMF')

ax[0].set_title(f'PMF: Sample vs Theory (r={sim_r}, p={sim_p})')
ax[0].legend()

# step 4: plot sample ECDF vs theoretical CDF
# plot the ECDF of the sample
sns.ecdfplot(sample, ax=ax[1], label='Sample ECDF', color='blue')

# overlay the theoretical CDF
theoretical_cdf = nbinom.cdf(k=k_vals, n=sim_r, p=sim_p)
ax[1].step(k_vals, theoretical_cdf, where='post', color= 'red', linestyle='--', label = 'Theoretical CDF')

ax[1].set_title(f'CDF: Sample vs Theory (r={sim_r}, p={sim_p})')
ax[1].legend()

plt.show()
```

### Categorical (similar to Bernoulli): The probability mass function over $k = 1, 2, ..., K$ categories is
$$
f(k;p_1,...,p_K) = p_1^{k=1}p_2^{k=2}...p_K^{k=K}
$$
where $0 \le p_i \le 1$ and $ \sum_{k=1}^K p_k = 1$.

This is like a generalization of the Bernoulli coin flip to something like a $K$ sided die where the sides don't necessarily have equal probabilities.

#### 6.4a Determine the support

The support is the set of categories, $K$. (The values $k$ can take)

#### 6.4b Plot the density and distribution for a variety of parameter values
